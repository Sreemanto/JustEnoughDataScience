```python
import numpy as np
```

```python
#normal function
def squreroot(l1): #parameter is a list of integers
    l2 = []
    for i in l1:
        l2.append(np.sqrt(i))
    return l2
```

```python
list1 = [64,225,400,144]
squreroot(list1)
```

```python
sqrt = lambda x:np.sqrt(x)
```

```python
list(map(sqrt,list1))
```

```python
list(map(squreroot,list1))
```
