```python
# Write a function to find square root of each element in a list
#normal function
def squareroot(l1):
    l2 = []
    for i in l1:
        l2.append(np.sqrt(i))
    return l2
```

```python
list1 = [64,225,36,100,400]
squareroot(list1)
```

```python
#with lambda
sqrt = lambda x:np.sqrt(x)

list(map(sqrt,list1))
list(map(lambda x:np.sqrt(x),list1))
```
