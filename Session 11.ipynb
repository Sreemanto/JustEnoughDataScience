```python
#Import the libraries and packages we need
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#we are going to create two arrays that have our data. These should be the same length

x = np.array([10,40,25,30,55,18,7]).reshape(-1,1) #we have to use reshape because we are required to pass in x 
                                                  #as a 2-dimensional array. One column and as many rows as necessary
y = np.array([5,18,22,45,60,34,19])
```

```python
x
```

```python
y
```

```python
print(x.shape)
print(y.shape)
```

```python
#we now create our model instance. We are just going to use the default parameters
#we also are going to fit the model
#this calculates the values for the slope and intercept in our regression equation
model = LinearRegression()
model.fit(x,y)
```

```python
#we can also find our intercept and slope
intercept = model.intercept_
slope = model.coef_
print("intercept: ",intercept)
print("slope: ",slope)
```

```python
#we can now create our regression equation
y_predicted = intercept + slope*x
y_predicted
```

```python
#now we can plot the 
fig,ax = plt.subplots(figsize=(10,10))
plt.scatter(x,y)
plt.plot(x,y_predicted)
plt.show()
```

#### A real life use case

```python
df = pd.read_csv('Premier_League_Table.csv')
df
```

```python
#Simple Feature Engineering
df['GD'] = df['GF'] - df['GA']
df
```

```python
y=df.Rk
x=df.GD.values.reshape(-1,1)
```

```python
x.shape
```

```python
y.shape
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
intercept = model.intercept_
slope = model.coef_
print(intercept)
print(slope)
```

```python
#Eqn of the line
y_pred = intercept + slope*x
y_pred
```

```python
# First object fig, short for figure, imagine it as the frame of your plot. 
#You can resize, reshape the frame but you cannot draw on it.

#The second object, ax, short for axes, is the canvas you draw on. 
#Or rephrasing, it is the blank sheet you can plot and hold your data. An axes object can only belong to one figure.
fig,ax = plt.subplots(figsize=(10,10))
fig.set_facecolor('wheat')
ax.patch.set_facecolor('azure')

#plot data
plt.scatter(x,y)
plt.plot(x,y_pred,c='red',linestyle='--')
plt.ylim(0,21) # I want to have a specified limit for the range of y value
plt.gca().invert_yaxis() # I want the Ranked 1 team to be at top
plt.yticks(np.arange(1,21)) #fixing the point axis
plt.xlabel('Goal Differential',fontsize=18)
plt.ylabel('Final Leage Rank',fontsize=18)
plt.title('Evaluating the Relationship Between Goal Differential and League Table Rank\nPremier League Seasons 1995/96 - 2019/2020',fontsize=16)
plt.show()
#plt.savefig('premline.png)
```
